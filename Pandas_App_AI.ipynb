### Concating Data Frames

```python
import pandas as pd
India_weather = pd.DataFrame({
    "city":["Mumbai","Delhi","Bangalore"],
    "Tempreture":[32,45,30],
    "hummmidity":[80,60,78]})
India_weather
```

```python
Us_tempreture = pd.DataFrame({
   "City":["New York","Chicago","Orlando"],
    "Tempreture":[21,14,35],
    "Hummidity":[68,65,75]
})
Us_tempreture
```

### Concating

```python
df = pd.concat([India_weather,Us_tempreture],ignore_index = True)
df
```

```python
df = pd.concat([India_weather,Us_tempreture],axis = 1)
df
```

# Merge Data Frame (In SQL  it's join)

```python
import pandas as pd
Tempreture_df = pd.DataFrame({
    "city":["Mumbai","Delhi","Bangalore","Hyderabad"],
    "Tempreture":[32,45,30,75]
})
Tempreture_df
```

```python
import pandas as pd
hummidity_df = pd.DataFrame({
    "city":["Mumbai","Delhi","Bangalore"],
    "hummmidity":[80,60,78]})
hummidity_df
```

### Merge

```python
df = pd.merge(Tempreture_df,hummidity_df, on = 'city', how = 'left')
df
```
