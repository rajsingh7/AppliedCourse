# Q-Q Plot

```python
import numpy as np 
import pandas as pd
import scipy.stats as stats

#N(0,1)
std_normal = np.random.normal(loc = 0,scale=1,size=1000)  #loc = mean ,Scale = std-deviation. .normal == normal/Gaussian Distribution  

#0 to 100th percentile of std-normal 

for i in range(0,101):

    print(i,np.percentile(std_normal,i))
       
```

```python
import pylab
import matplotlib.pyplot as plt

#N(20,5)
measurements =  np.random.normal(loc = 20, scale=5,size=1000) # 

stats.probplot(measurements, dist="norm", plot =pylab) # dist =norm means i am comparing with normal dist N(0,1) to this 
pylab.show()


# note:- it's a staraing line so x and y same type of distributioon.
```

```python
#generate 100 sample data from 
measurements = np.random.uniform(low=-1,high=1,size = 50000)

stats.probplot(measurements,dist = "norm",plot = pylab)


pylab.show()


#x is uniform and y is normal/gaussian distribution Curve
```
