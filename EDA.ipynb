```python
# What about 1-D scatter plot using just one feature?
#1-D scatter plot of petal-length
import numpy as np
iris_setosa = data.loc[data["Species"] == "Iris-setosa"];
iris_virginica = data.loc[data["Species"] == "Iris-virginica"];
iris_versicolor = data.loc[data["Species"] == "Iris-versicolor"];
#print(iris_setosa["petal_length"])
plt.plot(iris_setosa["Petal_Length"], np.zeros_like(iris_setosa['Petal_Length']), 'o')
plt.plot(iris_versicolor["Petal_Length"], np.zeros_like(iris_versicolor['Petal_Length']), 'o')
plt.plot(iris_virginica["Petal_Length"], np.zeros_like(iris_virginica['Petal_Length']), 'o')

plt.show()
#Disadvantages of 1-D scatter plot: Very hard to make sense as points 
#are overlapping a lot.
#Are there better ways of visualizing 1-D scatter plots?
```

```python
import numpy as np
counts, bin_edges = np.histogram(iris_setosa['Petal_Length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
counts, bin_edges = np.histogram(iris_virginica['Petal_Length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

```

```python
import numpy as np
print("Mean:")
print(np.mean(iris_virginica['Petal_Length']),"-Virginica")
print(np.mean(np.append(iris_setosa['Petal_Length'],50)),"-setosa with outlayer")  #adding one outlier
print(np.mean(iris_setosa['Petal_Length']),"-setosa without outlier")
print(np.mean(iris_versicolor['Petal_Length']),"-versicolor")

print("\nStd-dev:")

print(np.std(iris_virginica['Petal_Length']),"-Virginica")
print(np.std(iris_setosa['Petal_Length']),"-setosa")
print(np.std(iris_versicolor['Petal_Length']),"-versicolor")



# iris_setosa = data.loc[data["Species"] == "Iris-setosa"];
# iris_virginica = data.loc[data["Species"] == "Iris-virginica"];
# iris_versicolor = d
```

```python
#median
print("\nMedian")
print(np.median(iris_virginica['Petal_Length']),"-Virginica")
print(np.median(np.append(iris_setosa['Petal_Length'],50)),"-setosa with outlayer")  #adding one outlier
print(np.median(iris_setosa['Petal_Length']),"-setosa withou outlier")
print(np.median(iris_versicolor['Petal_Length']),"-versicolor")


#Quantiles
print("\nQuantiles")

print(np.percentile(iris_virginica['Petal_Length'],np.arange(0,100,25)),"--->percentile of Virginica")
print(np.percentile(iris_setosa['Petal_Length'],np.arange(0,100,25)),"--->percentile of  setosa")
print(np.percentile(iris_versicolor['Petal_Length'],np.arange(0,100,25)),"--->percentile of versicolor")

#90th percentile

print("\n90th percentile")

print(np.percentile(iris_setosa['Petal_Length'],90),"--setosa")
print(np.percentile(iris_virginica['Petal_Length'],90),"--Virginica")
print(np.percentile(iris_versicolor['Petal_Length'],90),"--vercicolour")

from statsmodels import robust
print("\nMedian Absolute Deviation:")
print(robust.mad(iris_setosa['Petal_Length'])," --setosa")
print(robust.mad(iris_versicolor['Petal_Length']), "--Versicolor")
print(robust.mad(iris_virginica['Petal_Length']),  "--Virginica")
```
