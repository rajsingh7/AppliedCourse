```python
# What about 1-D scatter plot using just one feature?
#1-D scatter plot of petal-length
import numpy as np
survived = haberman.loc[haberman["status"] == 1];
not_survived = haberman.loc[haberman["status"] ==  2];

#print(iris_setosa["petal_length"])
plt.plot(survived["age"], np.zeros_like(survived['age']), 'o')
plt.plot(not_survived["nodes"], np.zeros_like(not_survived['nodes']), 'o')

plt.show()
#Disadvantages of 1-D scatter plot: Very hard to make sense as points 
#are overlapping a lot.
#Are there better ways of visualizing 1-D scatter plots?
```

```python
# CDF  and PDF
counts, bin_edges = np.histogram(survived['age'], bins=10, 
                                 density = True)
pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
counts,bin_edges = np.histogram(not_survived['nodes'],bins = 10,density = True)
pdf = counts/sum(counts)
print(pdf)
print(bin_edges)

#calculating CDF

cdf = np.cumsum(pdf)
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.plot()
```
